# Q3. Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.

```python
from scipy.stats import norm 
import numpy as np  
def pdf(data,mean,std):
    probability_pdf = norm.pdf(data, loc=mean, scale=std)
    return probability_pdf
array = np.linspace(10,50,20)
array = [ round(i,2) for i in array] 
mean = np.mean(array)
std = np.std(array)
ans = pdf(12,mean,std)
ans
```

# Q4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.

### The binomial distribution is a statistical model that can be used to predict the likelihood of an event with two possible outcomes, such as success or failure. 
### Some properties of the binomial distribution include:
### Fixed probability of success: The probability of success remains the same for each trial.
### Independent trials: The outcome of each trial is independent of the other.
### Two mutually exclusive outcomes: There are only two possible outcomes, such as success or failure. 
### for example tossing a coin , pass or fail.

# Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4 and plot a histogram of the results using matplotlib. 

```python
from scipy.stats import binom 
p = 0.4
n = 1000
r_values = list(range(n + 1)) 
mean, var = binom.stats(n, p) 
dist = [binom.pmf(r, n, p) for r in r_values ]  
print("r\tp(r)") 
for i in range(n + 1): 
	print(str(r_values[i]) + "\t" + str(dist[i])) 
print("mean = "+str(mean)) 
print("variance = "+str(var))
```

```python
from scipy.stats import binom 
import matplotlib.pyplot as plt  
n = 1000
p = 0.4
r_values = list(range(n + 1)) 
dist = [binom.pmf(r, n, p) for r in r_values ] 
plt.bar(r_values, dist) 
plt.show()
```

# Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution with given mean at a given point.

```python
from scipy.stats import poisson
import matplotlib.pyplot as plt
# mean = 3
x = poisson.rvs(mu=3, size=20)
print(x)
plt.hist(x, density=True, edgecolor='black')
```

```python
def cumulative(x):
    probab = x/x.sum()
    cdf = np.cumsum(probab)
    return cdf

orgcdf = cumulative(x)
orgcdf
```

# Q7. How Binomial distribution different from Poisson distribution?

### The main difference between the binomial and Poisson distributions is that the binomial distribution is used for experiments with a fixed number of trials, while the Poisson distribution is used for events that occur within a fixed interval
